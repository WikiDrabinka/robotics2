```python
from jetbot import Camera, bgr8_to_jpeg
import ipywidgets.widgets as widgets
import traitlets
```

```python
c = Camera()
```

```python
widget = widgets.Image(format='jpeg', width=500, height=500)
display(widget)
```

```python
traitlets.dlink((c, 'value'), (widget, 'value'), transform=bgr8_to_jpeg)
```
